```python
tr = stats.truncnorm(-3, 2)
tr.stats('mv')
```

```python
mean, std = 2, 4
ma, mb = 0, 4
a, b = (ma - mean)/std, (mb - mean)/std
x = np.linspace(-5, 5, num=1000)
# tr = stats.truncnorm(-1, 3, loc=1, scale=1)
# pdf = tr.pdf(x)
# pdf = stats.truncnorm.pdf(x, a, b, mean, std)
pdf = stats.truncnorm.pdf(x, a, b)
print(a, b)
fig = sns.lineplot(x=x, y=pdf)
fig.set_xlabel('X')
fig.set_ylabel('PDF')
```

```python
cdf = stats.norm.cdf(x, loc=0, scale=1)
fig = sns.lineplot(x=x, y=cdf)
fig.set_xlabel('X')
fig.set_ylabel('CDF')
```

PPF(Percent Point Function)  
令P(x<=0.1) = 0.5,则ppf(0.5)=0.1

```python
cdf_ = np.linspace(0, 1, num=1000)
x_ = stats.norm.ppf(cdf_, loc=0, scale=1)
fig = sns.lineplot(x=x_, y=cdf_)
fig.set_xlabel('X')
fig.set_ylabel('CDF')
```
